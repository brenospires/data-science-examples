# Loss Landscapes, Overparameterization, and Implicit Bias

---

## 1. What Is a Loss Landscape?

A loss landscape is the function:

$$
L(\theta)
$$

where:

- $\theta \in \mathbb{R}^P$ are the parameters of the model.
- $P$ can be millions in deep networks.

The landscape lives in parameter space, not input space.

Optimization corresponds to moving through this high-dimensional surface using gradient descent:

$$
\theta_{t+1} = \theta_t - \eta \nabla L(\theta_t)
$$

---

## 2. Convex vs Non-Convex Landscapes

### Linear Regression + MSE

$$
L(w) = \|Xw - y\|^2
$$

This is a quadratic function:

$$
L(w) = w^T X^T X w - 2 y^T X w + y^T y
$$

If $X^T X$ is positive definite:

- The loss is convex.
- There is a single global minimum.
- Any critical point is global.

Convexity implies:

$$
L(\lambda \theta_1 + (1-\lambda)\theta_2)
\le
\lambda L(\theta_1) + (1-\lambda)L(\theta_2)
$$

So no bad local minima exist.

---

### Logistic Regression

Loss:

$$
L(w) = \sum_i \log(1 + e^{-y_i w^T x_i})
$$

Even though the model is nonlinear in inputs, it is convex in $w$ because:

- $\log(1+e^z)$ is convex.
- Composition with linear function preserves convexity.

---

## 3. Why Hidden Layers Break Convexity

Consider a one-hidden-layer model:

$$
f(x) = w_2 \sigma(w_1 x)
$$

Parameters appear multiplicatively.

Example toy loss:

$$
L(w_1, w_2) = (w_1 w_2 - 1)^2
$$

Minimum occurs when:

$$
w_1 w_2 = 1
$$

This defines a hyperbola with infinitely many solutions.

The minimum is not a single point but a curved manifold.

This implies:

- Redundant parameterizations
- Flat directions
- Non-convex geometry

---

## 4. Critical Points and Hessian

At a critical point:

$$
\nabla L(\theta) = 0
$$

Classification of critical points using Hessian eigenvalues:

- All positive → local minimum
- All negative → local maximum
- Mixed signs → saddle point
- Some zero, none negative → flat minimum

In high dimensions:

- Having all eigenvalues positive is unlikely.
- Most critical points are saddle points.

---

## 5. Why Saddle Points Dominate in High Dimensions

If the Hessian has $P$ eigenvalues:

- For a strict minimum, all must be positive.
- As $P$ increases, probability of mixed signs increases.

Thus:

> High-dimensional non-convex landscapes are dominated by saddle points.

---

## 6. Overparameterization

If $P \gg n$:

- There are more parameters than constraints.
- Infinitely many solutions achieve zero training loss.

Redundancy implies:

There exist directions $v$ such that:

$$
L(\theta + \epsilon v) = L(\theta)
$$

These are flat directions.

Hessian has zero eigenvalues.

Minimum becomes a manifold instead of a point.

---

## 7. Why Overparameterization Makes Optimization Easier

Overparameterization:

- **Redundancy → Flat directions in parameter space**  
  If multiple parameter configurations represent the same function, there exist directions $v$ such that:
  $$
  L(\theta + \epsilon v) = L(\theta)
  $$
  **Example:** for the loss $L(w_1, w_2) = (w_1 w_2 - 1)^2$, all points satisfying $w_1 w_2 = 1$ achieve zero loss, forming a flat valley.

- **Flat directions → Removal of isolated bad local minima**  
  Instead of small isolated high-loss bowls, the landscape becomes dominated by wide connected valleys of low loss.

- **High dimensionality → Many descent directions**  
  In large parameter spaces, it is unlikely that curvature is positive in every direction.  
  There is almost always some direction where loss can decrease.

- **Many descent directions → Saddle points dominate**  
  With many eigenvalues in the Hessian, mixed signs become far more common than all-positive curvature.  
  Thus most critical points are saddles rather than bad local minima.

- **Abundant solutions → Most minima achieve low training loss**  
  When $P \gg n$, the model can interpolate the dataset in many different ways.  
  Therefore most local minima correspond to low or zero training error.

- **Wide valleys → Connected low-loss regions**  
  Different parameter solutions often lie within a single connected basin of low loss, meaning they can be joined by a path that keeps loss small.

Optimization becomes finding any point in a wide valley.

## 8. Flat vs Sharp Minima

Sharp minimum:
- Large Hessian eigenvalues
- Loss increases quickly when moving away

Flat minimum:
- Many small eigenvalues
- Loss changes slowly nearby

SGD prefers flat minima because:

- It introduces noise (mini-batching).
- Noise destabilizes sharp minima.
- Flat regions are stable under perturbations.

---

## 9. Norm, Margin, and Generalization

Linear classifier:

$$
f(x) = w^T x
$$

Geometric margin:

$$
\gamma = \min_i \frac{y_i w^T x_i}{\|w\|}
$$

Under constraint $y_i w^T x_i \ge 1$:

$$
\gamma = \frac{1}{\|w\|}
$$

So minimizing $\|w\|$ maximizes margin.

Large margin implies:

- Robustness to input perturbations
- Lower sensitivity
- Better generalization

---

## 10. Implicit Bias of Gradient Descent

In separable logistic regression:

- Weight norm grows to infinity.
- Direction converges to maximum margin solution.

Why?

Because points far from the boundary saturate sigmoid and stop contributing to gradient.

Only boundary points (support vectors) dominate.

Thus gradient descent implicitly maximizes margin.

---

## 11. Big Picture Summary

Deep networks are:

- Non-convex
- High-dimensional
- Highly overparameterized

Yet they train successfully because:

- Saddle points dominate instead of bad minima.
- Overparameterization creates flat valleys.
- SGD prefers flat / low-norm solutions.
- Low-norm solutions correspond to large margin and robustness.

Optimization and generalization are shaped not only by capacity, but by geometry.
