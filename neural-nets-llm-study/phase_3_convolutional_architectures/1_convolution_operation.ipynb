# Convolution Operation (1D and 2D)

---

# 1. Intuition

Convolution is a **sliding local dot product**.

Given:
- An input signal (or image) $x$
- A small pattern (kernel / filter) $w$

At each position, convolution:
1. Extracts a local region
2. Multiplies elementwise with the kernel
3. Sums the result

So convolution is:

$$
y[t] = \langle x_{\text{local}}, w \rangle
$$

It measures **local similarity between the input and the kernel**.

---

# 2. 1D Convolution (Discrete Case)

Mathematical definition:

$$
(y * w)[t] = \sum_{k} x[t-k] \, w[k]
$$

In deep learning, we typically compute cross-correlation:

$$
y[t] = \sum_{k} x[t+k] \, w[k]
$$

The kernel is learned, so flipping does not change the hypothesis class.

---

## Interpretation

Convolution computes a **local dot product**.

Using:

$$
\langle x, w \rangle = \|x\| \|w\| \cos\theta
$$

Convolution measures:
- Alignment (via $\cos\theta$)
- Strength (via norms)

It is **not cosine similarity**, because we do not normalize.

---

# 3. Geometric Interpretation

Each local patch is a vector in $\mathbb{R}^k$.
The kernel is also a vector in $\mathbb{R}^k$.

Convolution computes:

$$
y = \langle \text{patch}, w \rangle
$$

So a filter defines a **direction in vector space**.

If activation is large and positive:
- Patch and filter are aligned (angle ≈ 0°)

If activation is large and negative:
- They are opposite (angle ≈ 180°)

If activation ≈ 0:
- They are nearly orthogonal

---

# 4. 2D Convolution (Images)

For images:

$$
X \in \mathbb{R}^{H \times W}
$$

Kernel:

$$
W \in \mathbb{R}^{k \times k}
$$

At each spatial position:

$$
y(i,j) = \sum_{a,b} X_{i+a, j+b} \, W_{a,b}
$$

This is again a dot product between:

- A flattened $k \times k$ patch
- A flattened kernel

So a $3 \times 3$ patch lives in:

$$
\mathbb{R}^9
$$

If the image has $C$ channels:

$$
\text{Patch space} = \mathbb{R}^{k \times k \times C}
$$

---

# 5. Edge Detection Example

Kernel:

$$
\begin{bmatrix}
-1 & 0 & 1 \\
-1 & 0 & 1 \\
-1 & 0 & 1
\end{bmatrix}
$$

This computes:

$$
\text{(right side)} - \text{(left side)}
$$

It approximates a **horizontal derivative**:

$$
\frac{\partial I}{\partial x}
$$

It detects vertical edges:
- Dark → Bright → positive response
- Bright → Dark → negative response
- Uniform region → zero

---

# 6. Output Size Formula

For 1D convolution:

$$
\text{output size} =
\frac{n + 2p - k}{s} + 1
$$

Where:
- $n$ = input size
- $k$ = kernel size
- $p$ = padding
- $s$ = stride

Example:
- $n = 10$
- $k = 3$
- $s = 1$
- $p = 0$

$$
10 - 3 + 1 = 8
$$

---

## "Same" Convolution

To preserve size when $s = 1$:

$$
p = \frac{k-1}{2}
$$

For $k = 3$:

$$
p = 1
$$

---

# 7. Conceptual Summary

Convolution is:

- A sliding window operation
- A local dot product
- A projection onto a learned direction in patch space
- A feature detector
- A discrete spatial derivative (for edge filters)

Geometrically:

> A CNN layer projects image patches onto learned directions in high-dimensional space.

```python
import numpy as np

rng = np.random.default_rng()

img = rng.integers(low=0, high=10, size=(5, 5))
filter = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])

def conv_2d(X, W):
    H, W_img = X.shape
    k, _ = W.shape
    
    out_H = H - k + 1
    out_W = W_img - k + 1
    Y = np.zeros((out_H, out_W))
    
    for i in range(out_H):
        for j in range(out_W):
            patch = X[i, j]
            Y[i, j] = X[i, :] @ X[j, :]
            
    return Y

conv_2d(img, filter)
```
