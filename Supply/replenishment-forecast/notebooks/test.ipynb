### Two Sums

```python
def two_sum(nums, target):
    seen = {}

    for i, num in enumerate(nums):
        complement = target - num

        if complement in seen:
            return [seen[complement], i]

        seen[num] = i

    return []

nums = [2, 7, 11, 15]
target = 9

two_sum(nums, target)
```

### Valid Parentheses

```python


def valid_parentheses(str):

    stack = []

    pairs = {
        ')': '(',
        ']': '[',
        '}': '{'
    }

    for char in str:

        if char in "([{":
            stack.append(char)
        else:

            if len(stack) == 0:
                return False
            
            top = stack.pop()

            if top != pairs[char]:
                return False

    return len(stack) == 0

s = "([]{})"
valid_parentheses(s)
```

### Top K Frequent Elements

```python
nums = [1,1,1,2,2,3]
k = 2

frequencies = {}

for n in nums:

    if n in frequencies:
        frequencies[n] = frequencies[n] + 1
    else:
        frequencies[n] = 1

bigger = []

for i in range(k):

    aux_val = 0
    aux_key = 0

    for key, freq in frequencies.items():
    
        if aux_val < freq:
            aux_val = freq
            aux_key = key

    bigger.append(aux_key)
    del frequencies[aux_key]

bigger
```

### Merge Intervals

```python

def merge_intervals(intervals):
    intervals.sort()
    merged_intervals = []

    current_start = None
    current_end = None

    for start, end in intervals:

        if current_end is None:
            current_start = start
            current_end = end
            continue

        if start <= current_end:
            current_end = max(current_end, end)
        else:
            merged_intervals.append((current_start, current_end))
            current_start = start
            current_end = end

     if current_end is not None:
        merged_intervals.append((current_start, current_end))
    
    return merged_intervals

intervals = [
    (8, 10),
    (1, 3),
    (15, 18),
    (2, 6)
 ]

merge_intervals(intervals)
```

```python
class TreeNode:

    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

def dfs(node, levels):
    
    print(node.val, levels)

    if node.left is None and node.right is None:
        return levels, node.val
        print("stop")
    

    dfs(node.left, levels + 1)
    dfs(node.right, levels + 1)
    
dfs(root, 1)

def deepest_leftmost_value(root):
    levels, val = dfs(root, 1)
    print(levels, val)
   
```
